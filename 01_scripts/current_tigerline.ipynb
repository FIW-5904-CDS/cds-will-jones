```python
import os
import pygris
import pandas as pd
import geopandas as gpd
```

```python
print(os.getcwd())
```

```python
from pygris import roads_nation

# Download all roads in the USA
all_roads = roads_nation()
```

```python
nc_counties = pygris.counties(state= 37)
ky_counties = pygris.counties(state= 21)
wv_counties = pygris.counties(state= 54)
md_counties = pygris.counties(state= 24)
tn_counties = pygris.counties(state= 47)
dc_county = pygris.counties(state= 11)
va_county = pygris.counties(state=51)
```

```python
counties = pd.concat([nc_counties, ky_counties, wv_counties, md_counties, tn_counties, dc_county, va_county], ignore_index=True)
```

```python
type(counties)
```

```python
va_roads = gpd.GeoDataFrame()

for _, county in va_counties.iterrows():
    county_fips = county['COUNTYFP']
    state = county['STATEFP']
    roads = pygris.roads(state= state , county=county_fips)
    va_roads = gpd.GeoDataFrame(pd.concat([va_roads, roads], ignore_index=True))
```

```python
all_roads = gpd.GeoDataFrame()

for _, county in counties.iterrows():
    county_fips = county['COUNTYFP']
    state = county['STATEFP']
    roads = pygris.roads(state= state , county=county_fips)
    all_roads = gpd.GeoDataFrame(pd.concat([all_roads, roads], ignore_index=True))
```

```python
processed_counties = set()  # Keep track of completed counties

for _, county in counties.iterrows():
    county_fips = county['COUNTYFP']
    if county_fips in processed_counties:
        continue  # Skip already processed counties

    state = county['STATEFP']
    roads = pygris.roads(state=state, county=county_fips)
    all_roads = gpd.GeoDataFrame(pd.concat([all_roads, roads], ignore_index=True))

    # Add the county to the processed list and save progress
    processed_counties.add(county_fips)
    all_roads.to_file("roads_downloaded_so_far.geojson", driver="GeoJSON")
```
