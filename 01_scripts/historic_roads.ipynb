### Trying to automate download of old 2000 RT1 (ASCII) files from https://www2.census.gov/geo/tiger/tiger2k/. 

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time
```

```python
os.getcwd()
os.chdir(r'E:\01_PROJECTS\01_Roads\01_scripts\tiger2k')
os.getcwd()
```

```python
base_url = "https://www2.census.gov/geo/tiger/tiger2k/"
```

```python
download_root = "tiger2k"
os.makedirs(download_root, exist_ok=True)

# Get all state directories
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
for link in soup.find_all('a'):
    href = link.get('href')
    # Add null check here
    if href and href.endswith('dc/') and href != '../':
        state_dir = href
        state_url = urljoin(base_url, state_dir)
        print(f"\nEntering {state_url}")

        # Make local directory
        local_state_dir = os.path.join(state_dir.strip('/'))
        os.makedirs(local_state_dir, exist_ok=True)

        # Get files in state directory
        state_response = requests.get(state_url)
        state_soup = BeautifulSoup(state_response.text, 'html.parser')

        for file_link in state_soup.find_all('a'):
            file_name = file_link.get('href')
            # Add null check here too
            if file_name and file_name.endswith(('.zip')):
                file_url = urljoin(state_url, file_name)
                local_file_path = os.path.join(local_state_dir, file_name)

                if not os.path.exists(local_file_path):
                    print(f"  Downloading {file_name}")
                    file_response = requests.get(file_url)
                    with open(local_file_path, 'wb') as f:
                        f.write(file_response.content)
                    time.sleep(0.5)  # Be polite to the server
                else:
                    print(f"  Already exists: {file_name}")
```
